```python
from datetime import datetime as dt
import shutil
import glob
import os

def tempToPers():

  sources = ['crimes', 'prices', 'economicStatus']

  for source in sources:
  
    # Moving from temporal to persistent and adding the timestamp
    cFiles = glob.glob("../../data/landing/temporal/" + source + "/*.csv")
    for file in cFiles:
      month = os.path.basename(file).split("-")[0:2]
      month = "-".join(month)
      shutil.copy2(file, "../../data/landing/persistent/" + source + "/" +  source + month + "-t-" + dt.now().strftime("%Y-%m-%d-%H_%M_%S") + ".csv")
      os.remove(file)


tempToPers()
```
